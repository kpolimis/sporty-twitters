#Sporty Twitters

This project main objective is to find a correlation between the well-being of users and their physical activity. This project relies on Twitter data and machine learning algorithms. It is divided in x parts:

- build a well-being classifier;
- build a classifier able to identify users that regularly exercise;
- study the well-being of users who exercise and users who don't.

The goal of this documentation is to help users to reproduce our experiment. Each module use is described and methods' inputs, outputs and behaviours are explicated.

The scripts are developed in Python using modules. To use them, open a terminal and go to the `src` folder, then use the following command:

    python -m folder.script params [options] < input > output

To begin, we have to collect data using the Twitter API. To do so, we use the [python-twitter](https://github.com/bear/python-twitter) module and develop our own script in the `tweet_collect` module.

### tweet_collect

This module provides a method to collect tweets by filtering on strings to track. It is also possible to give a maximum number of tweets to collect. From a terminal, you can run the script using the following command:

    python -m tweet_collect.collect --settings SETTINGS_FILE --limit LIMIT track [track ...]

The settings file is a JSON file that contains the authentification key provided by Twitter. The format to use is:

    {
    "consumer_key" : "",
    "consumer_secret" : "",
    "access_token" : "",
    "access_token_secret" : ""
    }

A minimum of one string to track is required. It can be hashtags, usernames, names, etc.

```python
import tweet_collect.collect as collect

# authentification using a JSON settings file
api = collect.authenticate(open("../settings.json", "r"));
# let's track sports applications tweets
tags2track = ['runkeeper','nikeplus','runtastic','endomondo','mapmyrun','strava','cyclemeter','fitstats','mapmyfitness','runmeter']
# run the collect of two tweets
collect.collect(api, limit=2, track=tags2track)
```

Once we have a big amount of data, we have to preprocess it in order to remove some properties (stopwords, hyperlinks, ...), this is the role of the `clean_corpus` module.

### clean_corpus
This module provides a script to:

- remove stopwords
- remove mentions
- remove URLs
- remove punctuation

It takes a corpus as an input and process it line by line. The default behaviour is to remove everything listed above but flags exist to keep some properties. Moreover, it is possible to cancel stopwords removing by not giving a stopwords file.

From a terminal, the command to run the script is:

    python -m clean_corpus.clean_corpus [stopwords_file,--keep-mentions,--keep-urls,--keep-punctuation] < corpus > cleaned_corpus

```python
import clean_corpus.clean_corpus as clean_corpus

corpus_file = "../ipython_examples/20_raw_tweets"
stopwords = "../inputs/stopwords"
rm_mentions = True
rm_urls = True
rm_punctuation = True

# Showing content of the input
with open(corpus_file, "r") as f:
    pretty_out.print_header("Input:", over=True)
    pretty_out.pprint(f)

# Process the input and show the output
pretty_out.print_header("Output:", over=True)
output = StringIO.StringIO()
with open(corpus_file, "r") as f:
    clean_corpus.clean(f, output, stopwords, rm_urls, rm_mentions, rm_punctuation)
pretty_out.pprint(output.getvalue())
```

### context_similar

This module contains three python scripts aiming to find words that have a **similar context** (i.e. similar neighbourhood) to a given set of words.

- `build_contexts` takes a corpus file as an input and outputs the contexts in a JSON format.

		`python -m context_similar.build_contexts < corpus > contexts.json`
   A context is defined by the neighbourhood of a word. For each processed word P, a context is a dictionary that associates a pair (P, pos(W, P)) and the frequency of the word W, where pos(W,P) is the position of W relatively to P.
For example, in the string "hello how are you", we have:
    - pos("hello", "how") = -1
    - pos("you", "how") = 2
    - etc...

  To reduce the computation time, we only consider a neighbourhood of range 3 (i.e. only the words which position is in [-3,3]).

- `extend_poms` takes the contexts and the POMS words as inputs and outputs the `n` most similar words for each POMS category. It is also possible to restrict the computed POMS categories to a subset using optional parameters.

		`python -m context_similar.extend_poms poms_file poms_legend_file [n, categories] < contexts.json > similarWords`

- `similarity_scores` contains the implemented methods to compute similarity scores between two words given their contexts:
	- $cosineSimilarity(\vec{A}, \vec{B}) = \frac{\vec{A}.\vec{B}}{||\vec{A}||\times||\vec{B}||}$
    

```python
import context_similar.build_contexts as build_contexts
import json

# use a clean corpus as input
corpus_file = "../ipython_examples/20_cleaned_tweets"

output = StringIO.StringIO()
with open(corpus_file, "r") as f:
    build_contexts.run(f, output)

contextdict = json.loads(output.getvalue())
print json.dumps(contextdict, sort_keys=True, indent=4, separators=(',', ': '))
```
